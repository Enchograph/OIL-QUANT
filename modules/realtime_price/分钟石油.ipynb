```python
import time
import pandas as pd
from datetime import datetime
import os
import yfinance as yf

# ================= 终极代理配置区 =================
PROXY_URL = "http://127.0.0.1:7897" 

os.environ['http_proxy'] = PROXY_URL
os.environ['https_proxy'] = PROXY_URL
os.environ['HTTP_PROXY'] = PROXY_URL
os.environ['HTTPS_PROXY'] = PROXY_URL
# ================================================

# ================= 业务配置区 =================
SYMBOL = "CL=F"  
CSV_FILENAME = "oil_prices_log.csv"
# ==========================================

def fetch_and_log_price():
    # 每次刷新前清空屏幕，让它看起来像个专业的动态看板
    os.system('cls' if os.name == 'nt' else 'clear')
    print(f"[{datetime.now().strftime('%H:%M:%S')}] ➡️ 正在通过全局环境代理请求数据，请稍候...")
    
    try:
        ticker = yf.Ticker(SYMBOL)
        # ⚠️ 关键修复：改为 5d，完美避开 yfinance 在休市/无数据时报 NoneType 的底层 bug
        df = ticker.history(period="1d", interval="1m")
        
        if not df.empty:
            latest = df.iloc[-1]
            kline_time = df.index[-1].strftime('%Y-%m-%d %H:%M:%S')
            
            row_data = {
                "fetch_time": datetime.now().strftime('%Y-%m-%d %H:%M:%S'),
                "kline_time": kline_time,
                "symbol": SYMBOL,
                "open": float(latest["Open"]),
                "high": float(latest["High"]),
                "low": float(latest["Low"]),
                "close": float(latest["Close"])
            }
            
            # 后台悄悄保存到 CSV
            df_save = pd.DataFrame([row_data])
            file_exists = os.path.isfile(CSV_FILENAME)
            df_save.to_csv(CSV_FILENAME, mode='a', header=not file_exists, index=False)
            
            # =========== 📊 屏幕监控面板（展示 OHLC） ===========
            print(f"✅ 数据获取成功！")
            print(f"📡 监控目标: {SYMBOL} (1分钟线)")
            print(f"⏰ 抓取时间: {row_data['fetch_time']} | 📊 K线时间: {row_data['kline_time']}")
            print("=" * 45)
            print(f"🟢 开盘 (Open):  $ {row_data['open']:.2f}")
            print(f"📈 最高 (High):  $ {row_data['high']:.2f}")
            print(f"📉 最低 (Low):   $ {row_data['low']:.2f}")
            print(f"🔴 收盘/最新:    $ {row_data['close']:.2f}")
            print("=" * 45)
            print(f"💾 数据已自动追加至: {CSV_FILENAME}")
            print("提示: 按 Ctrl+C 停止监控。")
            
        else:
            print("⚠️ 成功连上，但未获取到有效数据。")
            print("-" * 45)
            
    except Exception as e:
        print(f"❌ 网络或处理异常: {e}")

print("🚀 启动雅虎财经 1 分钟级石油监控程序...")

try:
    while True:
        fetch_and_log_price()
        
        # 精确等待到下一个整分钟
        now = datetime.now()
        sleep_seconds = 60 - now.second - (now.microsecond / 1_000_000.0)
        time.sleep(sleep_seconds)
        
except KeyboardInterrupt:
    print("\n🛑 已停止监控，数据已安全保存在本地。")
```
